```python
import geopandas as gpd
import contextily as cx
```

```python
# Load the Metro Manila boundary .geojson file
mm_boundary = gpd.read_file('../graphs/geojson/manila_1195.geojson')
```

```python
mm_boundary.crs
```

```python
mm_boundary_mercator = mm_boundary.to_crs(epsg=3857)
```

```python
# Display the Metro Manila boundary on a map using contextily
ax = mm_boundary_mercator.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax=ax, url=cx.providers.CartoDB.Positron, zoom=15)
```
