```python
import osmnx as ox
```

## Generate Graph

```python
# bounding box for Cebu City (Mandaue Not Included)
NORTH_BBOX = 10.36035
SOUTH_BBOX = 10.25631
WEST_BBOX = 123.82828
EAST_BBOX = 123.98964
```

```python
# bounding box for Cebu City and Mandaue City
NORTH_BBOX = 10.4426
SOUTH_BBOX = 10.2345
WEST_BBOX = 123.7809
EAST_BBOX = 124.1036
```

```python
CEBU_BOUNDING_BOX = (NORTH_BBOX, SOUTH_BBOX, EAST_BBOX, WEST_BBOX)
```

```python
cebu_graph = ox.graph_from_bbox(bbox=CEBU_BOUNDING_BOX)
```

##Save Graph

```python
from datetime import datetime
current_date = datetime.utcnow()
print(current_date)
```

```python
# save graph as a geopackage file
ox.save_graph_geopackage(cebu_graph, f'graphs/geopackage/{current_date} graph.gpkg', directed=True)
```

```python
# save graph as a graph ml file
ox.save_graphml(cebu_graph, f'graphs/graphml/{current_date} graph.graphml')
```

## Plot Graph

```python
ox.plot_graph(cebu_graph, bgcolor='#ffffff', node_color='#000000', edge_color='#000000')
```
