```python
import osmnx as ox
```

## Generate Graph

```python
# bounding box for Cebu City (Mandaue Not Included)
NORTH_BBOX = 14.7156
SOUTH_BBOX = 14.4247
WEST_BBOX = 121.3095
EAST_BBOX = 120.9460
```

```python
METRO_MANILA_BOUNDING_BOX = (NORTH_BBOX, SOUTH_BBOX, EAST_BBOX, WEST_BBOX)
```

```python
metro_manila_graph = ox.graph_from_bbox(bbox=METRO_MANILA_BOUNDING_BOX)
```

##Save Graph

```python
import datetime
current_date = datetime.datetime.now(datetime.UTC)
# current_date = datetime.now().strf time('%Y-%m-%d_%H-%M-%S')
print(current_date)
```

```python
# save graph as a geopackage file
ox.save_graph_geopackage(metro_manila_graph, f'C:/Users/Jed Padro/DataspellProjects/JeepneyRerouting/graphs/geopackage/{current_date} graph.gpkg', directed=True)
```

```python
# save graph as a graph ml file
ox.save_graphml(metro_manila_graph, f'C:/Users/Jed Padro/DataspellProjects/JeepneyRerouting/graphs/graphml/{current_date} graph.graphml')
```
